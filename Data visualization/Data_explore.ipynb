## Info data
This database was obtained from kaggle with the under the original name of "Global Forest Insights 1990-2020 | FAO". Made by Kanchana Karunarathna.

The original information includes:

Column Descriptors:
Country and Area: Name of the country or region.
Forest Area (1990-2020): Forest coverage in 1000 ha across different years.
Total Land Area (2020): Total land area in 1000 ha.
Forest Proportion (%): Forest area as a percentage of total land area.
Deforestation Rate: Change in forest area over time.
Area Affected by Fire: Forest area impacted by fires in 2015.

This dataset is ethically sourced from the United Nations' FAO Global Forest Resources Assessment 2020, ensuring credible and responsible data usage.

### *Which are the countries with the largest forest area?*

```python
df_nat.head(10)
```

```python
df_forest_area= df_nat.nlargest(8,"Forest Area, 2020")
df_forest_area
```

```python
# Total de deforestación en el mundo
plt.figure(figsize=(18, 6))

plt.plot(df_forest_area["Country and Area"], df_forest_area["Forest Area, 1990"], color='darkblue', linestyle='dotted', label='1990')
plt.plot(df_forest_area["Country and Area"], df_forest_area["Forest Area, 2010"], color='blueviolet', linestyle='dashdot', linewidth=1.5, label='2010')
plt.plot(df_forest_area["Country and Area"], df_forest_area["Forest Area, 2020"], color='seagreen', linestyle='dashed', label='2020')
plt.annotate('815311.60 is  \n the maximum value recorded.', xy=(0, 815311.60),fontsize=10, color='royalblue',xytext=(0.30, 790000), arrowprops=dict(arrowstyle= '-|>', color='royalblue'))

plt.title("Forest Area across diferent years")
plt.xlabel("Countries with the largest forest area")
plt.ylabel("Forest coverage in 1000 ha")
#plt.grid(True)

plt.legend(bbox_to_anchor=(1,.75))
plt.show()
```

```python
plt.figure(figsize=(18, 6))
forest_melted = pd.melt(df_forest_area, value_vars=['Forest Area, 1990', 'Forest Area, 2010', 'Forest Area, 2020'], var_name='Year', value_name='Forest_Area')
forest_melted.head(3)
sns.kdeplot(data=forest_melted, x='Forest_Area', hue='Year', fill=True)
plt.title("Density distribution")
```

### *What is the top 5 countries with the most deforested area?*

```python
df_mostdeforested= df_nat.nlargest(5,"Deforestation, 2015-2020")
df_mostdeforested
```

```python
fig = plt.figure(figsize = (8, 6), tight_layout = True, facecolor='aliceblue')
plt.bar(df_mostdeforested["Country and Area"], df_mostdeforested["Deforestation, 2015-2020"], color="slategrey", label='Deforestation, 2015-2020')
plt.title("Areas with the highest deforestation")
plt.xlabel('Country and Area')
plt.ylabel('Deforestation')
plt.annotate('1695.70', (0, 1710), fontsize=10, ha='center', weight='bold')
plt.annotate('668.40', (1, 680), fontsize=10, ha='center', weight='bold')
plt.annotate('650.00', (2, 675), fontsize=10, ha='center', weight='bold')
plt.annotate('474.00', (3, 490), fontsize=10, ha='center', weight='bold')
plt.annotate('293.92', (4, 310), fontsize=10, ha='center', weight='bold')

plt.tick_params(axis='x',labelrotation=20)
plt.legend(frameon=False)
plt.show()
```

### Which are the 5 countries with the least deforested area?

```python
df_fewerdeforested= df_nat[df_nat["Deforestation, 2015-2020"]>0].nsmallest(5,"Deforestation, 2015-2020")
df_fewerdeforested
```

```python
fig = plt.figure(figsize = (8, 6), tight_layout = True, facecolor='aliceblue')
fewdef = sns.barplot(data= df_fewerdeforested, x= 'Country and Area', y='Deforestation, 2015-2020', palette="GnBu", errorbar=None)
fewdef.set(title="Areas with less deforestation", xlabel="Año de registro", ylabel="Valores de registro")
#plt.axvline(x=36.5, ymin=0, ymax=1, linewidth=2,linestyle='--', color='red', label="Aumento de registros")
#plt.legend()
#plt.xticks(rotation=50)
fewdef.annotate('0.02', (0, 0.021), fontsize=10, ha='center', weight='bold')
fewdef.annotate('0.04', (1, 0.041), fontsize=10, ha='center', weight='bold')
fewdef.annotate('0.05', (2, 0.051), fontsize=10, ha='center', weight='bold')
fewdef.annotate('0.06', (3, 0.061), fontsize=10, ha='center', weight='bold')
fewdef.annotate('0.10', (4, 0.101), fontsize=10, ha='center', weight='bold')
plt.show()
```

### How much area is affected by fire in 2015 and which countries have these values?

```python
#df_fire= df_nat[df_nat["Total Forest Area Affected by Fire, 2015"]>0]
df_fire= df_nat.nlargest(10,"Total Forest Area Affected by Fire, 2015")
df_fire.reset_index(drop=True, inplace=True)
df_fire
```

```python
fig = plt.figure(figsize = (8, 6), tight_layout = True, facecolor='aliceblue')
#plt.plot(df_fire['Country and Area'], df_fire['Total Forest Area Affected by Fire, 2015'])
plt.pie(df_fire['Total Forest Area Affected by Fire, 2015'], labels=df_fire['Country and Area'], startangle=60, colors=['darkorange', 'gold', 'tomato', 'yellowgreen','darkmagenta',
               'teal', 'indigo',  'powderblue', 'royalblue'])
plt.title('10 countries most affected by fire in 2015')
plt.show()
```

### How is the ratio of forest area to total area in contrast to the average?

```python
average_proportion= df_nat["Forest Area as a Proportion of Total Land Area, 2020"].mean()
```

```python
df_nat["Forest Area as a Proportion of Total Land Area, 2020"].max()
```

```python
plt.hist(df_nat["Forest Area as a Proportion of Total Land Area, 2020"], bins=30, color='cadetblue')
plt.title('Lobsters tagged by size')
plt.ylabel('Value count')
plt.xlabel('Forest Area : Total Land Area')
plt.axvline(x=average_proportion, ymin=0, ymax=1, linewidth=2,linestyle='--', color='mediumblue')
plt.annotate('Average= 31.57 \nforest:land ratio, 2020', (34, 23), color='mediumblue')
#plt.annotate('Maximum legal catch size', (128, 7), color='mediumblue')
plt.show()
```
